数据列含义

age: 主要受益人年龄

sex: 保险合约人性别

bmi: 体重指数，提供了一个指标来衡量相对与身高来说体重是重了还是轻了

children: 有几个小孩

smoker: 是否吸烟

region: 地域

charges: 保费

```python
sns.scatterplot(x=data['bmi'], y=data['charges'])
```

这个散点图可以看出体重指数bmi和保费charges有很好的相关性，体重指数高的投保人倾向于买保额更高的保险，这个也很好理解，一般体重指数高的人群患病的风险也高

```python
#年龄与保费的关系
sns.scatterplot(x=data['age'], y=data['charges'])
```

```python
#回归量化 regplot
sns.regplot(x=data['bmi'], y=data['charges'])
```

```python
#带颜色分类的散点图
#是否吸烟对体重指数和保费的影响
```

```python
sns.scatterplot(x=data['bmi'], y=data['charges'], hue=data['smoker'])
```

吸烟的人群付的保费相对于不吸烟的人群要多

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=data)
```

吸烟者的回归线的斜度相对于非吸烟者要陡，代表吸烟者中体重指数和保费的相关性更强，在吸烟人群中体重指数越高的人保费付的越多
